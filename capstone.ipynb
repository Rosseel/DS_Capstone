# Introduction/Business Problem

My model is going to predict the severity of a traffic accident based upon data.
There are 2 ways in which this model can influence matters.
Pre-emptively and reactively.

1.Pre-emptively:
All cars that have access to this program will be able to convey the riskpercentage to the driver. And in response a driver can adjust his own driving actions accordingly.
Such a tremendous safety-feature will be a good selling point for any car manufacturer.
As well as for Insurance Agencies who wish to ensure that the people adhere to a low risk driving style.

2.Reactively:
In having access to such a model, the first responders will be able to act more accordingly by sending the appropriate amount of personnel and equipment to the situation.
Knowing the severity of an accident sooner, will give an idea of the ensuing length and time of a traffic jam.
This information in turn will allow in-place systems to reroute traffic faster and better.
